## 練習時間
試著使用 sklearn datasets 的其他資料集 (boston, ...)，來訓練自己的線性迴歸模型，並加上適當的正則化來觀察訓練情形。

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import Ridge, Lasso
import numpy as np
```

```python
##### Wine #####
### red data ###
boston = datasets.load_boston()
X = boston['data']
y = boston['target']

### train test split ###
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2)

### Boston Ridge ###

Ridge1 = Ridge()
alpha = [0.0001, 0.0005,0.001, 0.005, 0.01, 0.05, 0.1, 0.3, 0.5, 0.8, 1, 1.5, 2, 2.5,3, 3.5 ,4, 4.5 ,5, 5.5,6, 6.5,7, 7.5 ,8, 8.5,9,9.5,10]
Ridge_score = []
for i in alpha:
    Ridge1.set_params(alpha = i)
    Ridge_score.append([i, cross_val_score(Ridge1, X_train, y_train, cv=5).mean()])
print(np.array(Ridge_score))

##### 發現 alpha 設置 0.1 R^2 最高 #####
Ridge1.set_params(alpha = 0.1)
Ridge1.fit(X_train, y_train)
Ridge1.predict(X_test)
print('Score',Ridge1.score(X_test, y_test))
```

```python
### Boston Lasso ###
Lasso1 = Lasso()
alpha = [0, 0.0001, 0.0005,0.001, 0.005, 0.01, 0.05, 0.1, 0.3, 0.5, 0.8, 1, 1.5, 2, 2.5,3, 3.5 ,4, 4.5 ,5, 5.5,6, 6.5,7, 7.5 ,8, 8.5,9,9.5,10]
Lasso_score = []
for i in alpha:
    Lasso1.set_params(alpha = i)
    Lasso_score.append([i, cross_val_score(Lasso1, X_train, y_train, cv=5).mean()])
print(np.array(Lasso_score))

##### 發現 alpha 設置 0.001 R^2 最高 #####
Lasso1.set_params(alpha = 0.001)
Lasso1.fit(X_train, y_train)
Lasso1.predict(X_test)
print('Score',Lasso1.score(X_test, y_test))
```

```python
############# House data ######################
house = datasets.fetch_california_housing()
X = house['data']
y = house['target']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)

### Ridge ###
Ridge2 = Ridge()
alpha = [0,0.001, 0.005, 0.01, 0.05, 0.1, 0.3, 0.5, 0.8, 1, 1.5, 2, 2.5,3, 3.5 ,4, 4.5 ,5, 5.5,6, 6.5,7, 7.5 ,8, 8.5,9,9.5,10,20,50,100]
Ridge2_score = []
for i in alpha:
    Ridge2.set_params(alpha = i)
    Ridge2_score.append([i, cross_val_score(Ridge2, X_train, y_train, cv=100).mean()])
print(np.array(Ridge2_score))


##### 發現 alpha 設置 100 R^2 最高 #####
Ridge2.set_params(alpha = 100)
Ridge2.fit(X_train, y_train)
Ridge2.predict(X_test)
print('Score',Ridge2.score(X_test, y_test))
```
